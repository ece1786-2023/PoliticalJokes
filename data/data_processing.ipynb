```python
import tiktoken
import pandas as pd

def num_tokens_from_string(string: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding("cl100k_base")
    num_tokens = len(encoding.encode(string))
    return num_tokens

pd.set_option('display.max_colwidth', None)
df = pd.read_csv('political_jokes.csv')
df.insert(loc=1, column='Number of Tokens', value=df['Text'].apply(num_tokens_from_string))
df.sort_values(by=['Rating', 'Number of Tokens'], ascending=False, inplace=True, ignore_index=True)
df.to_csv('political_jokes_sorted.csv', index=False)
display(df)

df_generator = df.iloc[::2]
display(df_generator)
df_generator.to_csv('generator.csv', index=False)

df_evaluator = df.iloc[1::2]
display(df_evaluator)
df_evaluator.to_csv('evaluator.csv', index=False)
```
